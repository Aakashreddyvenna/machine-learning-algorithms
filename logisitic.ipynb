```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = [iris.target_names[i] for i in iris.target]

# Drop NaN values (if any)
df = df.dropna(subset=['species'])

# Ensure consistent data type
df['species'] = df['species'].astype(str)

# Remove rows where species is 'setosa'
df = df[df['species'] != 'setosa'].reset_index(drop=True)

df.head()
```

```python
df['species']=df['species'].map({'vercolor':0,'virginica':1})
```

```python
##splitt dataset intto independentt and dependent features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(model,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```
