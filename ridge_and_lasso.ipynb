```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
#independet and depenndent
x=dataset
y=df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
#standardizing the elements
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
#cross validatio
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
mse=cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
#prediction
regred=regression.predict(x_test)
```

```python
regred
```

```python
import seaborn as sns
sns.distplot(y_test-regred)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,regred)
```

```python
score
```

# ridge & Lasso####

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_reggression = Ridge()
```

```python
ridge_reggression
```

```python
parameters={'alpha':[1,2,5,10,20,30,40,50,70,80,90,100]}
ridgecv=GridSearchCV(ridge_reggression,parameters,scoring='neg_mean_squared_error',cv=10)
ridgecv.fit(x_train,y_train)
```

```python
print(ridgecv.best_params_)
print(ridgecv.best_score_)
```

```python
ridge_pred=ridgecv.predict(x_test)
```

```python
import seaborn as sns
sns.distplot(y_test-ridge_pred)
```

```python
score=r2_score(y_test,ridge_pred)
score
```

```python
from sklearn.linear_model import Lasso
```

```python
parameters={'alpha':[1,2,5,10,20,30,40,50,70,80,90,100]}
lassocv=GridSearchCV(Lasso(),parameters,scoring='neg_mean_squared_error',cv=10)
lassocv.fit(x_train,y_train)
```

```python
print(lassocv.best_params_)
print(lassocv.best_score_)
```

```python
lasso_pred=lassocv.predict(x_test)
```

```python
import seaborn as sns
sns.distplot(lasso_pred-y_test)
```
